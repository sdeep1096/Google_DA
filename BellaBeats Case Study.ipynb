```python
#We can also see that the Activity Date column is formatted incorrectly
data['ActivityDate']=pd.to_datetime(data['ActivityDate'])
data.info()
```

```python
#there are a total of 33 different observations as represented by nunique function for the Id column
#gathering descriptive statistics per unique user in the dataset
data.groupby('Id').agg(['mean','std','min','max','count'])
```

```python
#from the above descriptive statistics we can see majority of users have registered 0 steps taken in a session
#getting a more granulated data on minimum steps taken in a session
min_data=data.groupby('Id').min()
min_data
```

```python
#creating aplot of calories burned vs total steps taken
sns.scatterplot(x='TotalSteps',y='Calories',data=min_data,hue='Id')
```

```python
#Consolidating the Total Active Minutes in a single column
data.head()
```

```python
data['Total_Minutes']=data['VeryActiveMinutes']+data['FairlyActiveMinutes']+data['LightlyActiveMinutes']+data['SedentaryMinutes']
data.head()
```

```python
data['Active_Time']=data['VeryActiveMinutes']+data['FairlyActiveMinutes']+data['LightlyActiveMinutes']
data.head()
```

```python
data['Percentage_of_Active_Time']=100*data['Active_Time']/data['Total_Minutes']
data.head()
```

```python
data.groupby('Id')['Total_Minutes','Active_Time','Percentage_of_Active_Time'].mean()
```

```python
#Visual representation
sns.set_theme()
fig_dims = (10, 8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.kdeplot(data=data['Percentage_of_Active_Time'])
```

```python
sns.set_theme()
fig_dims = (10, 8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.kdeplot(data=data['Total_Minutes'])
```

```python
#Extracting columns of Interest
df=data[['TotalDistance','Active_Time','Calories']]
df.head()
```

```python
fig_dims = (10, 8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.scatterplot(x=df['Active_Time'], y=df['TotalDistance'], hue=df['Calories'], size=df['Calories']).set_title('Total Distance against Active Time')
```

```python
#Generating heatmap for correlation matrix
heatmap=sns.heatmap(df.corr(),annot=True,cmap='viridis')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12})
```
